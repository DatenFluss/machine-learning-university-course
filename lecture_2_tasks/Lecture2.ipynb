Import libraries needed for the task

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

load datasets and create graphs from the lecture

```python
# Load the datasets
iris_path = 'https://raw.githubusercontent.com/DatenFluss/machine-learning-university-course/main/res/iris.csv'
video_games_sales_path = 'https://raw.githubusercontent.com/DatenFluss/machine-learning-university-course/main/res/video_games_sales.csv'

iris_df = pd.read_csv(iris_path)
video_games_sales_df = pd.read_csv(video_games_sales_path)

# Drop NaN values from both datasets
iris_df_clean = iris_df.dropna()
video_games_sales_df_clean = video_games_sales_df.dropna()

# Set up subplots
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Iris dataset visualizations
# Bar Plot for Species count
iris_species_count = iris_df_clean['species'].value_counts()
axs[0, 0].bar(iris_species_count.index, iris_species_count.values, color='skyblue')
axs[0, 0].set_title('Iris Species Count')
axs[0, 0].set_ylabel('Count')
axs[0, 0].set_xlabel('Species')

# Scatter plot for sepal length vs petal length
axs[0, 1].scatter(iris_df_clean['sepal_length'], iris_df_clean['petal_length'], c='blue')
axs[0, 1].set_title('Sepal Length vs Petal Length')
axs[0, 1].set_xlabel('Sepal Length')
axs[0, 1].set_ylabel('Petal Length')

# Line Plot for sepal width variation
axs[1, 0].plot(np.arange(len(iris_df_clean)), iris_df_clean['sepal_width'], color='green')
axs[1, 0].set_title('Sepal Width Variation')
axs[1, 0].set_xlabel('Sample Index')
axs[1, 0].set_ylabel('Sepal Width')

# Video Games Sales visualizations
# Bar Plot for Genre count
vg_genre_count = video_games_sales_df_clean['Genre'].value_counts()
axs[1, 1].bar(vg_genre_count.index, vg_genre_count.values, color='purple')
axs[1, 1].set_title('Video Game Genre Distribution')
axs[1, 1].set_ylabel('Count')
axs[1, 1].set_xlabel('Genre')

# Pie Chart for platform distribution
platform_distribution = video_games_sales_df_clean['Platform'].value_counts()
axs[2, 0].pie(platform_distribution.values, labels=platform_distribution.index, autopct='%1.1f%%', startangle=90)
axs[2, 0].set_title('Platform Distribution')

# Line Plot for Global Sales over time
global_sales_by_year = video_games_sales_df_clean.groupby('Year_of_Release')['Global_Sales'].sum()
axs[2, 1].plot(global_sales_by_year.index, global_sales_by_year.values, color='red')
axs[2, 1].set_title('Global Sales Over Time')
axs[2, 1].set_xlabel('Year')
axs[2, 1].set_ylabel('Global Sales (in millions)')

plt.tight_layout()
plt.show()

# Seaborn scatterplot for Sepal Length vs Sepal Width colored by species
sns.scatterplot(data=iris_df_clean, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()

# Seaborn scatterplot for Petal Length vs Petal Width colored by species
sns.scatterplot(data=iris_df_clean, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width by Species')
plt.show()
```
